```python
from langchain_openai import ChatOpenAI
```

```python
llm = ChatOpenAI(model="gpt-4o")
```

```python
prompts = [
    {"role":"system", "content":"You are a JavaScript Developer ?"},
    {"role":"user", "content":"Difference b/w map and filter ?"},
]

res = llm.invoke(prompts)
print(res.content)
```

### Prompt Templates - Dynamic Prompts

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
```

```python
llm = ChatOpenAI(model="gpt-4o")
```

```python
prompts = ChatPromptTemplate.from_messages([
    {"role":"system", "content":"You are a tranlator and translate input in {language}"},
    {"role":"user", "content":"{query}"},
])
```

```python
final_prompts = prompts.format_messages(language = "Bangla", query= "I love Python, and JavaScript ")

res = llm.invoke(final_prompts)
print(res.content)
```
